```python
%%sql

SELECT InvoiceDate,
    BillingCountry,
    BillingState,
    MIN(Total) AS min_invoice, 
    MAX(Total) AS max_invoice, 
    AVG(Total) AS avg_invoice 
FROM invoices
GROUP BY BillingCountry, BillingState
LIMIT 10;
```

```python
#Remember that we are using the `MIN`, `MAX`, and `AVG` functions to aggregate values in the `Pct_managed_drinking_water_services` column as well as view values in the `Region` and `Sub_region` columns. 
#Therefore, we have to specify a grouping criteria using the `GROUP BY` clause.
```

```python
%%sql

SELECT DISTINCT BillingCountry,
    COUNT(DISTINCT(BillingState)) AS Number_of_States,
    ROUND(SUM(Total),4) AS rounded_suminvoice_by_country
FROM invoices 
GROUP BY BillingCountry
LIMIT 10;
```

```python
%%sql

SELECT 
    DISTINCT BillingCountry, 
    SUM(Total), 
    CONCAT( 
        SUBSTRING(IFNULL(UPPER(BillingCountry), 'UNKNOWN'),1,4), 
        SUBSTRING(IFNULL(BillingState, 'UNKNOWN'),1,4), 
        SUBSTRING(IFNULL(Total, 'UNKNOWN'),-1) 
    ) AS Country_invoice_id 
FROM 
    invoices
GROUP BY BillingCountry;
```

```python
%%sql
SELECT 
    Albumid,
    MediaTypeid,
    GenreId,
    Milliseconds,
    AVG(UnitPrice) OVER(PARTITION BY Albumid) AS avg_unitprice_fortrack
FROM 
    tracks
WHERE 
    Albumid = 3 
    AND Milliseconds < 300000
    AND UnitPrice < 1;
```

```python
### Implementing the same solution above using subqueries
```

```python
%%sql


SELECT 
    Albumid,
    MediaTypeid,
    GenreId,
    Milliseconds,
    avg_unitprice_fortrack
FROM (
    SELECT
        Albumid,
        MediaTypeid,
        GenreId,
        UnitPrice,
        Milliseconds,
        AVG(UnitPrice) OVER(PARTITION BY Albumid) AS avg_unitprice_fortrack
    FROM
        tracks
    WHERE
        Albumid = 3
) AS subquery
WHERE 
    Milliseconds < 300000
    AND UnitPrice < 1;
```

```python
### Implementing the same solution using Common Table Expressions (CTEs)
```

```python
%%sql
 
WITH track2 AS (
    SELECT 
        Albumid,
        MediaTypeid,
        GenreId,
        Milliseconds,
        UnitPrice,
        AVG(UnitPrice) OVER(PARTITION BY Albumid) AS avg_unitprice_fortrack  
    FROM 
        tracks
    WHERE 
        Albumid  = 3
)

SELECT 
    Albumid,
    MediaTypeid,
    GenreId,
    Milliseconds,
    avg_unitprice_fortrack
FROM 
    track2
WHERE 
    Milliseconds < 300000
    AND UnitPrice < 1;
```
